```python
from my_ml_lib.datasets._loaders import load_fashion_mnist
from my_ml_lib.model_selection._split import train_test_split , train_test_val_split
from my_ml_lib.linear_models.classification import LogisticRegression
from my_ml_lib.nn.losses import BinaryCrossEntropyLoss , CrossEntropyLoss
from my_ml_lib.nn.autograd import Value
from my_ml_lib.nn.modules.containers import Sequential
from my_ml_lib.nn import Linear
from my_ml_lib.nn.optim import SGD
from my_ml_lib.nn.modules.activations import ReLU
import numpy as np
import matplotlib.pyplot as plt
from my_ml_lib.preprocessing._gaussian import GaussianBasisFeatures
from my_ml_lib.preprocessing._data import StandardScaler
```

```python
X,y = load_fashion_mnist()
X_train , X_val,X_test , y_train , y_val, y_test = train_test_val_split(X , y)
X_train.shape , X_val.shape , y_train.shape , y_val.shape
```

```python
transform = StandardScaler()
transform.fit(X_train)
X_train = Value(transform.transform(X_train))
X_val = Value(transform.transform(X_val))
X_test = Value(transform.transform(X_test))



```

### 2. $\texttt{SoftMax Regression}$

```python
def softmax(X_train = X_train , y_train = y_train , lr = 0.1,in_features = 784 , out_features = 10 , max_iters = 300 , print_it = True):
    max_iters = max_iters

    softmax_layer = Linear(in_features = in_features , out_features = out_features , bias = True)
    loss = CrossEntropyLoss()

    model = Sequential(softmax_layer)
    optimizer = SGD(params = model.parameters() , lr = lr)

    losses = []
    for i in range(max_iters):
        
        optimizer.zero_grad()
        logits = model(X_train)
        loss_val = loss(logits , y_train)
        loss_val.backward()
        optimizer.step()

        if(i % 20 == 0 and print_it):
            print(f'iteration {i} -> CE Loss : {loss_val.data}')

        losses.append(loss_val.data)
    
    return model , loss ,losses


    
def softmax_score(model , X_val , y_val , loss , print_it):
    logits = model(X_val)
    acc = np.mean(np.argmax(logits.data , axis = 1) == y_val)
    if(print_it):
        print("The Validation Set accurcay is : " ,acc)
    return acc    


```

```python
model , loss , losses_1 = softmax(lr = 0.01)
score_1 = softmax_score(model , X_val , y_val , loss ,print_it = True)
```

```python
model , loss , losses_2 = softmax(lr = 0.05)
score_2 = softmax_score(model , X_val , y_val , loss , print_it = True)
```

```python
model , loss , losses_3 = softmax(lr = 0.1)
score_3 = softmax_score(model , X_val , y_val , loss , print_it = True)
```

```python
model , loss , losses_4 = softmax(lr = 0.5)
score_4 = softmax_score(model , X_val , y_val , loss , print_it = True)
```

```python
plt.figure(figsize=(12,5))
plt.title("Training Loss vs Epochs for different learning rates")
plt.plot(np.arange(len(losses_1)) + 1 , losses_1 , label = 'lr = 0.01')
plt.plot(np.arange(len(losses_2)) + 1 , losses_2 , label = 'lr = 0.05')
plt.plot(np.arange(len(losses_3)) + 1 , losses_3 , label = 'lr = 0.1')
plt.plot(np.arange(len(losses_4)) + 1 , losses_4 , label = 'lr = 0.5')
plt.grid()
plt.legend()
plt.show()
```

```python
best_model_softmax , loss , losses = softmax(lr = 0.1)
test_acc_softmax = softmax_score(best_model_softmax , X_test , y_test , loss , print_it = True)
print("The Test Accuracy for softmax is : " ,test_acc_softmax )
```

### $\texttt{SoftMax with Gaussian Features}$

```python
def train_softmax_with_gaussian(X_train , y_train ,X_val,y_val, lr  , n_centre , num_targets , max_iters , print_it):
    model , loss , losses = softmax(X_train = X_train ,y_train = y_train, lr = lr , in_features = n_centre , out_features = num_targets , max_iters = max_iters , print_it = print_it)
    score = softmax_score(model , X_val , y_val , loss , print_it)
    return losses , score
```

```python
n_centres_list = [ 200 , 300 ]
learning_rates = [0.01  , 0.1 , 0.5 , 1 , 2 , 4]

best_hp = []
best_score = 0.0
max_iters = 500

for n_centre in n_centres_list:

    gaussian = GaussianBasisFeatures(n_centers = n_centre)
    X_gauss = gaussian.fit_transform(X)
    X_tg , X_vg , y_tg , y_vg = train_test_split(X_gauss , y)
    X_tg = Value(X_tg)
    X_vg = Value(X_vg)
    for lr in learning_rates:
        print(f'Running for learning rate : {lr} and num_centres : {n_centre}')
        if(len(best_hp) == 0):
            best_hp = [n_centre , lr]
            best_score = train_softmax_with_gaussian(X_tg , y_tg , X_vg , y_vg , lr , n_centre, 10 , max_iters , False)[1]
        else:
            score = train_softmax_with_gaussian(X_tg , y_tg , X_vg , y_vg , lr , n_centre ,10 , max_iters , False)[1]
            if(best_score < score):
                best_hp = [n_centre , lr]
                best_score = score


```

```python
best_n_centres = best_hp[0]
best_lr = best_hp[1]

gaussian = GaussianBasisFeatures(n_centers = best_n_centres)
X_gauss = gaussian.fit_transform(X)
X_tg , X_vg , y_tg , y_vg = train_test_split(X_gauss , y)
X_tg = Value(X_tg)
X_vg = Value(X_vg)

losses , score = train_softmax_with_gaussian(X_tg , y_tg , X_vg , y_vg , best_lr , best_n_centres , 10 , max_iters , False )

```

### $\texttt{MLP}$

```python
max_iters = 75

layer1 = Linear(784 , 256)
activation_1 = ReLU()
layer2 = Linear(256 , 128)
activation_2 = ReLU()
layer3 = Linear(128 , 10)

losses_1 = []


model = Sequential(layer1 , activation_1 ,layer2 ,activation_2, layer3)

loss = CrossEntropyLoss()

optimizer = SGD(model.parameters() , lr = 0.01)



for i in range(max_iters):
    
    optimizer.zero_grad()
    logits = model(X_train)
    loss_val = loss(logits , y_train)
    losses_1.append(loss_val.data)
    loss_val.backward()
    optimizer.step()

    if(i % 25 == 0):
        print(f'iteration {i} -> CE Loss : {loss_val.data}')

train_acc = acc = np.mean(np.argmax(logits.data , axis = 1) == y_train)
print("Training accuracy : ",train_acc)

logits = model(X_val)
loss_val = loss(logits , y_val)
test_acc = np.mean(np.argmax(logits.data , axis = 1) == y_val)
print("Validation accuracy : ",test_acc)
```
